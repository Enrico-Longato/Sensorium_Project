```python
MOUSES_DATA_PATH = "/orfeo/scratch/mdmc/vnkanang/Sensorium/project_data/"
MOUSES_DATA_FOLDERS = [
    os.path.join(MOUSES_DATA_PATH, folder)
    for folder in os.listdir(MOUSES_DATA_PATH)
    if not folder.startswith("__MACOSX")
    and os.path.isdir(os.path.join(MOUSES_DATA_PATH, folder))
]

for mouse_data_folder in MOUSES_DATA_FOLDERS:
    meta_folder = os.path.join(mouse_data_folder, "meta")
    neurons_folder = os.path.join(meta_folder, "neurons")
    cell_motor_coordinates_path = os.path.join(neurons_folder, "cell_motor_coordinates.npy")
    if os.path.exists(cell_motor_coordinates_path):
        cell_motor_coordinates = np.load(cell_motor_coordinates_path)
        print(f"Mouse folder: {mouse_data_folder}, Cell motor coordinates shape: {cell_motor_coordinates.shape}")
    else:
        print(f"Mouse folder: {mouse_data_folder}, Cell motor coordinates file not found.")
    # 3d plot of the cell motor coordinates
    fig = plt.figure()
    ax = fig.add_subplot(111, projection='3d')
    ax.scatter(cell_motor_coordinates[:, 0], cell_motor_coordinates[:, 1], cell_motor_coordinates[:, 2])
    ax.set_xlabel('X Label')
    ax.set_ylabel('Y Label')
    ax.set_zlabel('Z Label')
    plt.show()
```
